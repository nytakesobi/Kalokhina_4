Гістограма відвідуваності за місяцями кожного музею. Вивести 4 графіки: за 2014, 2015, 2016, 2017 роки. Графіки розташовані два зверху, два знизу, колір першого графіка - синій, другого зелений, третього жовтий, четвертого сірий.

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
filtered_df = df[~df['Year'].str.contains('2018')]

museums = ["avila_adobe","firehouse_museum","chinese_american_museum","america_tropical_interpretive_center"]
unique_years = filtered_df['Year'].unique().tolist()
plt.figure(figsize=(12, 8))
subplt = 1
for museum in museums:
    museum_data = filtered_df[['Year', 'Month_Name', museum]]

    plt.subplot(2, 2, subplt)
    subplt = subplt + 1
    for year in unique_years:
        year_data = museum_data[museum_data['Year'] == year]
        plt.plot(year_data['Month_Name'], year_data[museum], marker='o', label=f'{year}')

    plt.title(f'{museum} Visits Over the Years')
    plt.xlabel('Month')
    plt.ylabel('Visits')
    plt.xticks(rotation=45)
    plt.legend(title='Year')

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
filtered_df = df[~df['Year'].str.contains('2018')]
colors = ['blue', 'green', 'yellow','grey']
museums = ["avila_adobe","firehouse_museum","chinese_american_museum","america_tropical_interpretive_center"]
unique_years = filtered_df['Year'].unique().tolist()
plt.figure(figsize=(12, 8))
subplt = 1
for year in unique_years:
    year_data = filtered_df[filtered_df['Year'] == year]

    plt.subplot(2, 2, subplt)

    for index, museum in enumerate(museums):
      plt.hist(year_data['Month_Name'], bins=12,color=colors[index], alpha=0.5, label=museum, edgecolor='black', align='left', weights=year_data[museum])

    subplt = subplt + 1

    plt.xticks(rotation=45)
    plt.legend(title='Year')

plt.tight_layout()
plt.show()
```
